## 画像の読み込み

画像ファイルを読み込む

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 画像ファイルを保存したgoogleドライブのフォルダの指定
# 自分の環境に合わせて書き換えること
dir_path = '/content/drive/MyDrive/OpenCV/'

# 画像ファイルの読み込み
img = cv2.imread(dir_path + "Pepper.jpg")

# 配列のサイズを画面に表示
print(img.shape)
```

読み込んだ画像を画面に表示する

```python
plt.imshow(img)
```

色情報の保存の仕方をBGRからRGBに変更する

```python
# BGRからRGBに変換
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

# 画像の回転

時計回りに90度回転させる

```python
img_rotate1 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(img_rotate1)
```

反時計回りに90度回転させる

```python
img_rotate2 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.imshow(img_rotate2)
```

180度回転させる

```python
img_rotate3 = cv2.rotate(img, cv2.ROTATE_180)

plt.imshow(img_rotate3)
```

## 画像の反転

上下に反転させる

```python
img_flip1 = cv2.flip(img, 0)

plt.imshow(img_flip1)
```

左右に反転させる

```python
img_flip2 = cv2.flip(img, 1)

plt.imshow(img_flip2)
```

上下にも左右にも反転させる

```python
img_flip3 = cv2.flip(img, -1)

plt.imshow(img_flip3)
```

## 画像のトリミング

配列の範囲を指定して画像の一部分をトリミングする

```python
img_trim = img[50:150, 50:130]

plt.imshow(img_trim)
```

## 画像の拡大、縮小

画素数300 x 300に拡大する

```python
img_large = cv2.resize(img, (300, 300))

plt.imshow(img_large)
```

画素数70 x 70に縮小する

```python
img_small = cv2.resize(img, (70, 70))

plt.imshow(img_small)
```
